```python
'''
Title:Web Scrapping Project: Sports Analytics for America’s National Football League
Author:Jedidah Wavinya
Date: 14 Ocober 2025
'''
#Importing libraries that I will need for webscraping
from bs4 import BeautifulSoup
import requests
import pandas as pd
import requests
```

```python
#set URL of website in a variable
url = 'https://www.nfl.com/stats/team-stats/'
page = requests.get(url)
```

```python
#Use BeautifulSoup to extract the HTML content
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Extract the table with the Team data
table = soup.find_all('table')

# Use pandas to read the HTML table into a DataFrame
if table:
    df = pd.read_html(str(table))[0]

    # Display the entire DataFrame
    display(df)
else:
    print("No table found on the page.")
```

Save the data in to a csv file

```python
df.to_csv('nfl_team_stats.csv', index=False)
```
